```python
# %% [markdown]
# # Ordenamiento y An√°lisis B√°sico de Datos CSV
# 
# Este notebook ordena los archivos CSV por columnas clave y muestra an√°lisis b√°sicos.

# %%
import pandas as pd
from IPython.display import display, HTML

# Configuraci√≥n de visualizaci√≥n
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)

# Lista de archivos CSV
archivos = {
    'alumnos': 'csv/alumnos.csv',
    'carreras': 'csv/carreras.csv',
    'grupos': 'csv/grupos.csv',
    'encuestas': 'csv/encuestas.csv',
    'users': 'csv/users.csv',
    'roles': 'csv/roles.csv',
    'migrations': 'csv/migrations.csv',
    'predicciones': 'csv/predicciones.csv'
}

# %% [markdown]
# ## Funci√≥n para ordenar datos

# %%
def ordenar_datos(nombre_archivo, ruta_archivo):
    print(f"\n{'='*80}")
    print(f"üìä ORDENAMIENTO: {nombre_archivo.upper()}")
    print('='*80)
    
    try:
        # Leer archivo
        df = pd.read_csv(ruta_archivo)
        
        # Informaci√≥n b√°sica
        print(f"\nüìå Total de filas: {len(df):,}")
        print(f"üìå Total de columnas: {len(df.columns)}")
        
        # Columnas disponibles
        print("\nüîç Columnas disponibles:")
        print(df.columns.tolist())
        
        # %% [markdown]
        ### Ordenamientos b√°sicos
        
        # %%
        # 1. Ordenamiento por ID (si existe)
        if '_id' in df.columns:
            print(f"\nüÜî Primeros registros ordenados por _id:")
            display(HTML(df.sort_values('_id').head().to_html()))
        
        # 2. Ordenamiento por fechas (si existen)
        fecha_cols = [col for col in df.columns if 'fecha' in col.lower() or 'date' in col.lower() or 'created' in col.lower() or 'updated' in col.lower()]
        if fecha_cols:
            print(f"\nüïí √öltimos registros ordenados por {fecha_cols[0]}:")
            display(HTML(df.sort_values(fecha_cols[0], ascending=False).head().to_html()))
        
        # 3. Ordenamiento por columnas clave espec√≠ficas
        if nombre_archivo == 'alumnos':
            if 'nombre_alumno' in df.columns:
                print("\nüßë Ordenamiento alfab√©tico por nombre:")
                display(HTML(df.sort_values(['app', 'apm', 'nombre_alumno']).head().to_html()))
            
            if 'id_carrera' in df.columns:
                print("\nüéì Conteo de alumnos por carrera:")
                display(HTML(df['id_carrera'].value_counts().to_frame().to_html()))
        
        elif nombre_archivo == 'encuestas':
            if 'correo' in df.columns:
                print("\nüìß Ordenamiento por correo electr√≥nico:")
                display(HTML(df.sort_values('correo').head().to_html()))
            
            if 'id_grupo' in df.columns:
                print("\nüë• Conteo de encuestas por grupo:")
                display(HTML(df['id_grupo'].value_counts().to_frame().to_html()))
        
        elif nombre_archivo == 'predicciones':
            if 'riesgo' in df.columns:
                print("\n‚ö†Ô∏è Distribuci√≥n de niveles de riesgo:")
                display(HTML(df['riesgo'].value_counts().to_frame().to_html()))
                
                print("\nüîç Predicciones ordenadas por riesgo:")
                display(HTML(df.sort_values('riesgo').head().to_html()))
        
        return df
        
    except Exception as e:
        print(f"\n‚ùå ERROR al procesar {nombre_archivo}: {str(e)}")
        return None

# %% [markdown]
# ## Ejecutar ordenamiento para todos los archivos

# %%
for nombre, ruta in archivos.items():
    ordenar_datos(nombre, ruta)

# %% [markdown]
# ## Funci√≥n para b√∫squeda simple

# %%
def buscar_valor():
    print("\n" + "="*80)
    print("üîç BUSCAR VALOR EN ARCHIVOS")
    print("="*80)
    
    archivo = input("¬øEn qu√© archivo quieres buscar? (alumnos/encuestas/predicciones): ").lower()
    columna = input("¬øEn qu√© columna quieres buscar?: ")
    valor = input("¬øQu√© valor quieres buscar?: ")
    
    try:
        df = pd.read_csv(archivos[archivo])
        resultados = df[df[columna].astype(str).str.contains(valor, case=False, na=False)]
        
        if not resultados.empty:
            print(f"\n‚úÖ Se encontraron {len(resultados)} coincidencias:")
            display(HTML(resultados.to_html()))
        else:
            print("\nüîç No se encontraron coincidencias.")
    except Exception as e:
        print(f"\n‚ùå Error: {str(e)}")
```
